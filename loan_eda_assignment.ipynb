### Converting `emp_length` (Employee Continuity Data) from String to Float  

- The column **`emp_length`** represents how many years a borrower has been employed.  
- It is stored as a **string** (e.g., `"10+ years"`, `"< 1 year"`, `"n/a"`).  
- For analysis, we convert it into a **numeric format (float)**:  
  - `"10+ years"` → `10`  
  - `"< 1 year"` → `0.5`  
  - `"n/a"` → `NaN` (missing)  
- This makes it easier to analyze the effect of employment length on loan default.  


```python
df["emp_length"]=df["emp_length"].str.extract(r"(\d+)")
counts=df["emp_length"].value_counts(normalize=True)
n_nulls = df["emp_length"].isna().sum()

fill_values = np.random.choice(counts.index, size=n_nulls, p=counts.values)

df.loc[df["emp_length"].isna(), "emp_length"] = fill_values
df["emp_length"] = df["emp_length"].astype(float)

df["emp_length"].isnull().sum()
```

### Converting `revol_util` from String to Float  

- The column **`revol_util` (Revolving Line Utilization Rate)** is stored as a string with a percentage sign (e.g., `"45%"`).  
- For numerical analysis, we need to **remove the `%` sign** and convert it to a **float type**.  
- This allows us to use it in calculations, statistical analysis, and visualizations.  


```python
df["revol_util"]=df["revol_util"].str.strip("%").astype("float")
df["revol_util"]=df["revol_util"].fillna(df["revol_util"].mean())
df["revol_util"].isnull().any()
```

### Final Step in Dropping Null Values  

- After performing various cleaning steps, some columns/rows may still contain **missing values**.  
- At this stage, we:  
  - Drop rows with very few remaining null values (since they form a small portion of data).  
  - Ensure the dataset has **no null values left** for smooth analysis and modeling.  
- This step finalizes the dataset for **feature engineering and visualization**.

### Data Type Conversion and Filtering
